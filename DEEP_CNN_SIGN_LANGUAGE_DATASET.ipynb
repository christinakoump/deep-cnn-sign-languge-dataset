3. **Διαχωρισμός Δεδομένων**: Το σύνολο δεδομένων διαχωρίζεται σε σύνολα εκπαίδευσης και δοκιμής (train/test) με ποσοστό 80% για εκπαίδευση και 20% για δοκιμή, χρησιμοποιώντας τη συνάρτηση train_test_split της βιβλιοθήκης sklearn.

4. **Αποκλεισμός Κρυμμένων Κλάσεων**: Επιλέγονται τυχαία δύο κλάσεις που θα αποκλειστούν από το σύνολο εκπαίδευσης. Αυτές οι κλάσεις δεν θα χρησιμοποιηθούν για την εκπαίδευση του μοντέλου, αλλά θα περιλαμβάνονται στο σύνολο δοκιμής για να μετρηθεί η απόδοση του μοντέλου στις κρυμμένες κλάσεις.

5. **Δημιουργία Μοντέλου Νευρωνικού Δικτύου**: Ορίζεται ένα CNN μοντέλο με τρία στρώματα συνελικτικών (convolutional) και ανώτατα στρώματα (max-pooling). Το μοντέλο περιλαμβάνει επίσης ένα βήμα ενίσχυσης των δεδομένων μέσω τυχαίων περιστροφών, ζουμ και οριζόντιας αναστροφής των εικόνων.

6. **Εκπαίδευση του Μοντέλου**: Το μοντέλο εκπαιδεύεται χρησιμοποιώντας τα δεδομένα εκπαίδευσης που δεν περιλαμβάνουν τις κρυμμένες κλάσεις. Χρησιμοποιούνται δύο callbacks: το EarlyStopping για να σταματήσει η εκπαίδευση αν η απόδοση δεν βελτιώνεται για 10 συνεχόμενες εποχές και το ReduceLROnPlateau για να μειώσει το learning rate αν η απόδοση δεν βελτιώνεται για 5 εποχές.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;--- Data Augmentation: Εφαρμόζει τυχαίες μετατροπές (π.χ. περιστροφή, ζουμ, αναστροφή) στις εικόνες κατά τη διάρκεια της εκπαίδευσης για να βοηθήσει το μοντέλο να μάθει &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;καλύτερα και να μην εξαρτάται από συγκεκριμένα χαρακτηριστικά.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;--- Conv2D: Εφαρμόζει φίλτρα στις εικόνες για να ανιχνεύσει χαρακτηριστικά όπως γωνίες, άκρα ή υφές. Αυτό γίνεται μέσω συνελικτικών φίλτρων (3x3) και της συνάρτησης ReLU για να &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;μάθει σύνθετα χαρακτηριστικά.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;--- MaxPooling2D: Μειώνει την ανάλυση των εικόνων.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;--- Conv2D:  Δεύτερο στρώμα συνελικτικών φίλτρων, το οποίο εξάγει πιο σύνθετα χαρακτηριστικά από τα πρώτα στρώματα.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;--- MaxPooling2D (Δεύτερη Μείωση Ανάλυσης): Μειώνει ακόμα περισσότερο τη διάσταση των χαρακτηριστικών για να διευκολύνει την επεξεργασία.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;--- Flatten: Μετατρέπει τα πολυδιάστατα δεδομένα από τα προηγούμενα στρώματα σε μονοδιάστατο πίνακα, για να μπορεί να τα επεξεργαστεί το επόμενο πλήρως συνδεδεμένο &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;στρώμα.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;--- Dense (Πλήρως Συνδεδεμένο Στρώμα): Το πρώτο πλήρως συνδεδεμένο στρώμα, το οποίο έχει 256 νευρώνες και χρησιμοποιεί ReLU για μη γραμμικότητα.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;--- Dense (Δεύτερο Πλήρως Συνδεδεμένο Στρώμα): Ένα δεύτερο πλήρως συνδεδεμένο στρώμα με 512 νευρώνες, επίσης με ReLU.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;--- Dense (Έξοδος): Το τελικό στρώμα που προβλέπει την κατηγορία της εικόνας με την ενεργοποίηση softmax για πολλαπλές κατηγορίες.

7. **Αξιολόγηση Μοντέλου**: Αφού ολοκληρωθεί η εκπαίδευση, το μοντέλο αξιολογείται τόσο στο σύνολο δοκιμής (που περιλαμβάνει και τις κρυμμένες κλάσεις) όσο και στις κρυμμένες κλάσεις μόνο.

8. **Απεικόνιση Εκπαίδευσης**: Στη συνέχεια, τα διαγράμματα της απώλειας (loss) και της ακρίβειας (accuracy) τόσο για την εκπαίδευση όσο και για την επικύρωση (validation) παρουσιάζονται, για να δείξουν την πρόοδο του μοντέλου κατά τη διάρκεια των εποχών.

### ΚΩΔΙΚΑΣ ΜΕ ΑΠΟΚΡΥΨΗ 2 ΚΛΑΣΕΩΝ 

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from tensorflow.keras import layers
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Load the dataset 
x_data = np.load('C:/Users/hp-5c/Desktop/deep_learning/archive/x.npy') #images
y_data = np.load('C:/Users/hp-5c/Desktop/deep_learning/archive/y.npy') #labels

# Print shapes
print("x_data shape:", x_data.shape)  
print("y_data shape:", y_data.shape)  

# Get the unique classes
classes = np.unique(y_data)
print("Names of classes:", classes)

# Define label dictionary for one-hot encoding
label_dict = {label: i for i, label in enumerate(classes)}
print("Labels matching:", label_dict)

# Map the class names to integers using the label_dict dictionary
y_int = np.array([label_dict[label] for label in y_data.flatten()])

# Convert the integer labels to one-hot encoded vectors
y_cat = to_categorical(y_int, num_classes=len(label_dict)).astype(int)

# Split data into train and test sets
x_train, x_test, y_train, y_test = train_test_split(x_data, y_cat, test_size=0.2, random_state=42)

# --- Exclude some classes from training ---
# Choose a random number of classes to hide (e.g., 2 classes)
hidden_classes = np.random.choice(classes, 2, replace=False)
print(f"Classes to be hidden during training: {hidden_classes}")

# Mask the training and test sets to exclude the selected hidden classes
train_mask = np.isin(np.argmax(y_train, axis=1), hidden_classes, invert=True)
test_mask = np.isin(np.argmax(y_test, axis=1), hidden_classes, invert=True)

# Update training and test sets
x_train_hidden = x_train[train_mask]
y_train_hidden = y_train[train_mask]
x_test_hidden = x_test[test_mask]
y_test_hidden = y_test[test_mask]

# Data augmentation for training
data_augmentation = keras.Sequential([
    keras.layers.RandomRotation(0.1),
    keras.layers.RandomZoom(0.1),
    keras.layers.RandomFlip("horizontal"),
])

# Build the model
model = keras.Sequential([
    data_augmentation,
    keras.layers.Conv2D(64, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    keras.layers.MaxPooling2D((2, 2)),
    keras.layers.Conv2D(128, (3, 3), activation='relu'),
    keras.layers.MaxPooling2D((2, 2)),
    keras.layers.Conv2D(256, (3, 3), activation='relu'),
    keras.layers.MaxPooling2D((2, 2)),
    keras.layers.Flatten(),
    keras.layers.Dense(256, activation='relu'),
    keras.layers.Dense(512, activation='relu'),
    keras.layers.Dense(len(label_dict), activation='softmax')
])

# Compile the model
model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Define callbacks
early_stopping = keras.callbacks.EarlyStopping(patience=10, restore_best_weights=True)
reduce_lr = keras.callbacks.ReduceLROnPlateau(factor=0.1, patience=5)

# Train the model on the data excluding the hidden classes
history = model.fit(x_train_hidden, y_train_hidden, epochs=10, validation_data=(x_test_hidden, y_test_hidden), callbacks=[early_stopping, reduce_lr])

# ------------------------- PREDICTION AND EVALUATION -------------------------
# After training, evaluate the model on the full test set (including hidden classes)
predictions_probs = model.predict(x_test)
predictions = np.argmax(predictions_probs, axis=1)
test_labels = np.argmax(y_test, axis=1)

# Accuracy on all classes
test_accuracy = accuracy_score(test_labels, predictions)
print("The Accuracy of the model on the full Test Data (including hidden classes) is", test_accuracy)

# Evaluate with model.evaluate() as well
loss, accuracy = model.evaluate(x_test, y_test)
print('Test accuracy (via model.evaluate):', accuracy)

# ------------------------------ PLOTTING TRAINING -----------------------------
# Plot the training and validation accuracy and loss
fig, axis = plt.subplots(nrows=1, ncols=2, figsize=(20, 6))

# Loss plot
axis[0].plot(history.epoch, history.history['loss'])
axis[0].plot(history.epoch, history.history['val_loss'])
axis[0].set_xlabel("Epochs")
axis[0].set_ylabel("Loss")
axis[0].legend(["Loss", "Val_loss"])
axis[0].set_title("Training Process - Loss")

# Accuracy plot
axis[1].plot(history.epoch, history.history['accuracy'])
axis[1].plot(history.epoch, history.history['val_accuracy'])
axis[1].set_xlabel("Epochs")
axis[1].set_ylabel("Accuracy")
axis[1].legend(["Accuracy", "Val_Accuracy"])
axis[1].set_title("Training Process - Accuracy")

plt.show()
```

### SELF- SUPERVISED METHOD SimCLR CODE

```python
# Load libraries
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
from tensorflow.keras import layers

# -------------------------- LOAD AND PREPROCESS DATA --------------------------

# Load the dataset 
x_data = np.load('C:/Users/hp-5c/Desktop/deep_learning/archive/x.npy')
y_data = np.load('C:/Users/hp-5c/Desktop/deep_learning/archive/y.npy')

# Normalize image data
x_data = x_data.astype("float32") / 255.0

# Print shapes
print("x_data shape:", x_data.shape)
print("y_data shape:", y_data.shape)

# Label encoding
y_labels = y_data.flatten()
classes = np.unique(y_labels)
label_dict = {label: i for i, label in enumerate(classes)}
y_int = np.array([label_dict[label] for label in y_labels])
y_cat = to_categorical(y_int, num_classes=len(label_dict)).astype(int)

# Train-test split
x_train, x_test, y_train, y_test = train_test_split(x_data, y_cat, test_size=0.2, random_state=42)

# ------------------------- SIMCLR PRETRAINING -------------------------

# Strong augmentations for SimCLR
simclr_augment = keras.Sequential([
    layers.Resizing(160, 160),  # Resize up before cropping
    layers.RandomCrop(128, 128),
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.2),
    layers.RandomZoom(0.2),
    layers.RandomContrast(0.2),
])

def create_simclr_pairs(x):
    return simclr_augment(x), simclr_augment(x)

def simclr_dataset(x_data, batch_size=128):
    dataset = tf.data.Dataset.from_tensor_slices(x_data)
    dataset = dataset.shuffle(1000).map(lambda x: create_simclr_pairs(x), num_parallel_calls=tf.data.AUTOTUNE)
    return dataset.batch(batch_size).prefetch(tf.data.AUTOTUNE)

pretrain_ds = simclr_dataset(x_data)

# Encoder network
def get_encoder():
    inputs = keras.Input(shape=(128, 128, 3))
    x = layers.Conv2D(64, 3, activation='relu')(inputs)
    x = layers.MaxPooling2D()(x)
    x = layers.Conv2D(128, 3, activation='relu')(x)
    x = layers.MaxPooling2D()(x)
    x = layers.Conv2D(256, 3, activation='relu')(x)
    x = layers.GlobalAveragePooling2D()(x)
    return keras.Model(inputs, x, name="encoder")

# SimCLR model = encoder + projection head
def get_simclr_model(encoder):
    inputs = keras.Input(shape=(128, 128, 3))
    features = encoder(inputs)
    x = layers.Dense(256, activation='relu')(features)
    outputs = layers.Dense(128)(x)
    return keras.Model(inputs, outputs, name="simclr_model")

encoder = get_encoder()
simclr_model = get_simclr_model(encoder)

# Contrastive loss (NT-Xent)
def nt_xent_loss(z_i, z_j, temperature=0.5):
    z_i = tf.math.l2_normalize(z_i, axis=1)
    z_j = tf.math.l2_normalize(z_j, axis=1)
    representations = tf.concat([z_i, z_j], axis=0)
    similarity_matrix = tf.matmul(representations, representations, transpose_b=True)

    logits_mask = tf.linalg.diag(tf.zeros(tf.shape(similarity_matrix)[0])) + 1e-9
    logits = similarity_matrix / temperature
    logits = logits * (1.0 - logits_mask)

    batch_size = tf.shape(z_i)[0]
    labels = tf.range(batch_size)
    labels = tf.concat([labels, labels], axis=0)

    loss = tf.keras.losses.sparse_categorical_crossentropy(labels, logits, from_logits=True)
    return tf.reduce_mean(loss)

# Pretraining loop
optimizer = keras.optimizers.Adam()
@tf.function
def train_step(x1, x2):
    with tf.GradientTape() as tape:
        z_i = simclr_model(x1, training=True)
        z_j = simclr_model(x2, training=True)
        loss = nt_xent_loss(z_i, z_j)
    gradients = tape.gradient(loss, simclr_model.trainable_variables)
    optimizer.apply_gradients(zip(gradients, simclr_model.trainable_variables))
    return loss

print("Starting SimCLR Pretraining...")
for epoch in range(10):
    total_loss = 0
    batches = 0
    for x1, x2 in pretrain_ds:
        loss = train_step(x1, x2)
        total_loss += loss
        batches += 1
    print(f"Epoch {epoch+1}, Loss: {total_loss / batches:.4f}")

# ------------------------- FINE-TUNING CLASSIFIER -------------------------

print("\nFine-tuning classifier on labeled data...")

classifier = keras.Sequential([
    encoder,
    layers.Dense(256, activation='relu'),
    layers.Dense(128, activation='relu'),
    layers.Dense(len(label_dict), activation='softmax')
])

classifier.compile(optimizer='adam',
                   loss='categorical_crossentropy',
                   metrics=['accuracy'])

early_stopping = keras.callbacks.EarlyStopping(patience=10, restore_best_weights=True)
reduce_lr = keras.callbacks.ReduceLROnPlateau(factor=0.1, patience=5)

history = classifier.fit(x_train, y_train, epochs=10,
                         validation_data=(x_test, y_test),
                         callbacks=[early_stopping, reduce_lr])

# ------------------------- EVALUATION -------------------------

predictions_nn_probs = classifier.predict(x_test)
predictions_nn = np.argmax(predictions_nn_probs, axis=1)
test_labels_multiclass = np.argmax(y_test, axis=1)

accuracy = accuracy_score(test_labels_multiclass, predictions_nn)
print("Test Accuracy:", accuracy)

# Plot training curves
fig, axis = plt.subplots(1, 2, figsize=(20, 6))
axis[0].plot(history.epoch, history.history['loss'])
axis[0].plot(history.epoch, history.history['val_loss'])
axis[0].set_title("Loss")
axis[0].legend(['Train', 'Val'])

axis[1].plot(history.epoch, history.history['accuracy'])
axis[1].plot(history.epoch, history.history['val_accuracy'])
axis[1].set_title("Accuracy")
axis[1].legend(['Train', 'Val'])

plt.show()

# Freeze encoder (no further updates during fine-tuning)
encoder.trainable = False

# Build full model for classification
finetune_model = keras.Sequential([
    encoder,                              # pretrained encoder from SimCLR
    layers.Flatten(),
    layers.Dense(256, activation='relu'),
    layers.Dense(512, activation='relu'),
    layers.Dense(len(label_dict), activation='softmax')  # output layer
])

# Redefine callbacks if you're in a new cell or scope
early_stopping = keras.callbacks.EarlyStopping(patience=10, restore_best_weights=True)
reduce_lr = keras.callbacks.ReduceLROnPlateau(factor=0.1, patience=5)

finetune_model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.001),
                       loss='categorical_crossentropy',
                       metrics=['accuracy'])

# You can reuse early_stopping and reduce_lr from your original code
history_finetune = finetune_model.fit(
    x_train, y_train,
    validation_data=(x_test, y_test),
    epochs=10,
    callbacks=[early_stopping, reduce_lr]
)
```
